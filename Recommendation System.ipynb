```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.neighbors import NearestNeighbors

# Create the user-item matrix
user_item_matrix = np.array([[5, 3, 0, 1, 4],
                             [0, 2, 3, 5, 1],
                             [1, 0, 5, 2, 4],
                             [4, 0, 2, 0, 0]])

# Create a dataframe for the movies
movies = pd.DataFrame({'title': ['A', 'B', 'C', 'D', 'E']})

# Print the user-item matrix and movies dataframe
print('User-Item Matrix:')
print(user_item_matrix)
print('\nMovies:')
print(movies)

# Compute item-item similarity using cosine similarity
item_sim_matrix = cosine_similarity(user_item_matrix.T)
np.fill_diagonal(item_sim_matrix, 0)

# Use k-nearest neighbors to find the most similar items
knn = NearestNeighbors(metric='cosine', algorithm='brute')
knn.fit(user_item_matrix.T)


def item_based_recommendations(item_id, knn=knn, item_sim_matrix=item_sim_matrix, movies=movies, k=2):
  _, indices = knn.kneighbors(item_sim_matrix[item_id].reshape(1, -1), n_neighbors=k+1)
  similar_items = indices[0][1:]
  recommended_items = [movies.iloc[i]['title'] for i in similar_items]
  return recommended_items

# Make item-based recommendations for item 0
recommendations = item_based_recommendations(0)
print(f'Item-based recommendations for item 0: {recommendations}')

# Use k-nearest neighbors to find the most similar users
knn = NearestNeighbors(metric='cosine', algorithm='brute')
knn.fit(user_item_matrix)

# Make recommendations based on the ratings of similar users
def user_based_recommendations(user_id, knn=knn, user_item_matrix=user_item_matrix, movies=movies, k=2):
    _, indices = knn.kneighbors(user_item_matrix[user_id].reshape(1, -1), n_neighbors=k+1)
    similar_users = indices[0][1:]
    ratings = user_item_matrix[similar_users].mean(axis=0)
    top_items = ratings.argsort()[-k:][::-1]
    recommended_items = [movies.iloc[i]['title'] for i in top_items]
    return recommended_items

# Make user-based recommendations for user 1
recommendations = user_based_recommendations(1)
print(f'User-based recommendations for user 1: {recommendations}')
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import mean_squared_error
from math import sqrt
import pandas as pd

# create a small dataset for demo purposes
data = {
    'User': ['User 1', 'User 2', 'User 3', 'User 4', 'User 5'],
    'Movie 1': [4, 3, 4, 2, 5],
    'Movie 2': [3, 4, 2, 5, 3],
    'Movie 3': [5, 2, 3, 4, 4],
    'Movie 4': [2, 5, 4, 3, 2],
    'Movie 5': [1, 4, 5, 2, 3]
}

df = pd.DataFrame(data)
print(df)

# calculate the mean ratings for each movie
mean_ratings = df.drop('User', axis=1).mean()

# normalize the ratings for each user by subtracting the mean rating
df_norm = df.drop('User', axis=1).apply(lambda x: x - mean_ratings[x.name], axis=1)

# calculate the cosine similarity between the users
user_similarity = cosine_similarity(df_norm.T)

# get the top n similar users to a given user
def get_similar_users(user_index, n=5):
    return user_similarity[user_index].argsort()[-n-1:-1][::-1]

# predict the rating for a movie by a user using collaborative filtering
def predict_rating(user_index, movie_index):
    # get the top 5 similar users to the given user
    similar_users = get_similar_users(user_index, n=5)
    similar_users = get_similar_users(user_index, n=5)
    similar_ratings = df_norm.iloc[similar_users[:-1], movie_index]



    # get the ratings of the similar users for the given movie
    similar_ratings = df_norm.iloc[similar_users.tolist(), movie_index]


    # calculate the predicted rating as the weighted average of the similar ratings
    weight = user_similarity[user_index][similar_users]
    prediction = (similar_ratings * weight).sum() / weight.sum()

    # add back the mean rating to get the final predicted rating
    prediction = prediction + mean_ratings[movie_index]

    return prediction

# example of predicting the rating for Movie 1 by User 1
prediction = predict_rating(0, 0)
print('Predicted rating for Movie 1 by User 1:', prediction)

# calculate the root mean squared error (RMSE) of the predictions
def calculate_rmse(predictions, actual_ratings):
    return sqrt(mean_squared_error(predictions, actual_ratings))

# example of calculating the RMSE of the predictions for all movies by all users
predicted_ratings = df_norm.apply(lambda row: row.index.map(lambda col: predict_rating(row.name, col)), axis=1)
actual_ratings = df_norm.copy()
rmse = calculate_rmse(predicted_ratings.values.flatten(), actual_ratings.values.flatten())
print('RMSE:', rmse)
```
