# ***Function for converting documents to shingles***

```python
def convert(file_name,nmin=1,nmax=1):
  doc = docx.Document(file_name)
  str1 = ' '
  for para in doc.paragraphs:
      str1 += para.text
  str1 = str1.replace(" ", "")
  str1 = str1.replace(",","")
  str1 = str1.replace(".","")
  str1 = str1.replace("'","")
  str1 = str1.replace("“","")
  shingles = ks.shingleset_range(str1, nmin, nmax)
  #shingles = ks.shingleseqs_range(str1,n_min = nmin,n_max = nmax)
  return shingles
```

# ***K-SHINGLES***

```python
k = int(input("Enter the number of shingles:"))
doc1 = convert('1.docx',k,k)
print(doc1)
doc2 = convert('2.docx',k,k)
print(doc2)
doc3 = convert('3.docx',k,k)
print(doc3)
doc4 = convert('4.docx',k,k)
print(doc4)
doc5 = convert('5.docx',k,k)
print(doc5)
```

## To calculate the Jaccard similarity between both, we take the number of shared words / total words and there is our Jaccard similarity - we can rewrite this into a function.

```python
def jac(x: set, y: set):
    shared = x.intersection(y)  # selects shared tokens only
    return len(shared) / len(x.union(y))  # union adds both sets together
```

```python
master = [doc1, doc2, doc3, doc4, doc5]
comb = list(combinations([0,1,2,3,4], 2))
for i in comb:
  print("Similarity of doc" + str(i[0]+1) + " and doc" + str(i[1]+1) + " = " + str(jac(master[i[0]], master[i[1]])))
```
